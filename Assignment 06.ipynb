```python
# Task 1
class Student:
    id = 0
    def __init__(self, name, dept, age, cgpa):
        self.name = name
        self.dept = dept
        self.age = age
        self.cgpa = cgpa
        Student.id += 1

    def get_details(self):
        print("ID:", Student.id)
        print("Name:", self.name)
        print("Department:", self.dept)
        print("Age:", self.age)
        print("CGPA:", self.cgpa)

    @classmethod
    def from_String(cls, string1):
        name, dept, age, cgpa = string1.split("-")
        obj = cls(name, dept, age, cgpa)
        return obj

s1 = Student("Samin", "CSE", 21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.get_details()

#subtask 5
print("Instance variables are non-static variables which are owned by instances of the class. Class variables are static variables that are shared with all instances of the class.")

#subtask 6
print("The class method takes a different parameter 'cls' instead of 'self' that points to the class instead of the object.")
```

```python
# Task 2
class Assassin:
    number = 0
    def __init__(self, name, success):
        self.name = name
        self.success = success
        Assassin.number += 1

    def printDetails(self):
        print("Name:", self.name)
        print(f"Success rate: {self.success}%")
        print("Total number of Assassin:", Assassin.number)

    @classmethod
    def failureRate(cls, name, failure):
        success = 100 - failure
        obj = cls(name, success)
        return obj
    @classmethod
    def failurePercentage(cls, name, fail_percent):
        failure = fail_percent.replace("%", "")
        success = 100 - int(failure)
        obj = cls(name, success)
        return obj

john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage("Akabane", "10%")
akabane.printDetails()
```

```python
# Task 3
class Passenger:
    count = 0
    def __init__(self, name):
        self.name = name
        self.fare = 450
        Passenger.count += 1

    def set_bag_weight(self, weight):
        if 21 <= weight <= 50:
            self.fare += 50
        elif weight > 50:
            self.fare += 100

    def printDetail(self):
        print("Name:", self.name)
        print("Bus Fare:", self.fare)


print("Total Passenger:", Passenger.count)
p1 = Passenger("Jack")
p1.set_bag_weight(90)
p2 = Passenger("Carol")
p2.set_bag_weight(10)
p3 = Passenger("Mike")
p3.set_bag_weight(25)
print("=========================")
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================")
print("Total Passenger:", Passenger.count)
```

```python
# Task 4
class Travel:
    count = 0
    def __init__(self, loc1, loc2, time = "1:00"):
        self.loc1 = loc1
        self.loc2 = loc2
        self.time = time
        Travel.count += 1

    def display_travel_info(self):
        str1 = f"Source: {self.loc1}\n"
        str1 += f"Destination: {self.loc2}\n"
        if str(self.time)[-3:] != ":00":
            str1 += f"Flight Time: {self.time}:00"
        else:
            str1 += f"Flight Time: {self.time}"

        return str1

    def set_time(self, new_time):
        self.time = new_time

    def set_destination(self, new_loc2):
        self.loc2 = new_loc2
    
    def set_source(self, new_loc1):
        self.loc1 = new_loc1

print("No. of Traveller =", Travel.count)
print("=======================")
t1 = Travel("Dhaka","India")
print(t1.display_travel_info())
print("=======================")
t2 = Travel("Kuala Lampur","Dhaka")
t2.set_time(23)
print(t2.display_travel_info())
print("=======================")
t3 = Travel("Dhaka","New_Zealand")
t3.set_time(15)
t3.set_destination("Germany")
print(t3.display_travel_info())
print("=======================")
t4 = Travel("Dhaka","India")
t4.set_time(9)
t4.set_source("Malaysia")
t4.set_destination("Canada")
print(t4.display_travel_info())
print("=======================")
print("No. of Traveller =", Travel.count)
```

```python
#Task 5
class Employee:
    def __init__(self, name, year):
        self.name = name
        self.workingPeriod = year

    @classmethod
    def employeeByJoiningYear(cls, name, join_year):
        year  = 2021 - join_year
        return cls(name, year)
    
    @staticmethod
    def experienceCheck(period, gender):
        if period < 3 and gender == "male":
            return "He is not experienced"
        elif period < 3 and gender == "female":
            return "She is not experienced"
        elif period >= 3 and gender == "female":
            return "She is experienced"
        elif period >= 3 and gender == "male":
            return "He is experienced"

employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
# Task 6
class Laptop:
    laptopCount = 0
    def __init__(self, name, count):
        self.name = name
        self.count = count
        Laptop.laptopCount += self.count

    @classmethod
    def advantage(cls):
        print("Laptops are portable")

    @classmethod
    def resetCount(cls):
        Laptop.laptopCount = 0
        
lenovo = Laptop("Lenovo", 5)
dell = Laptop("Dell", 7)
print(lenovo.name, lenovo.count)
print(dell.name, dell.count)
print("Total number of Laptops", Laptop.laptopCount)
Laptop.advantage()
Laptop.resetCount()
print("Total number of Laptops", Laptop.laptopCount)
```

```python
# Task 7
class Cat:
    Number_of_cats = 0
    def __init__(self, color, action):
        self.color = color
        self.action = action
        Cat.Number_of_cats += 1

    def printCat(self):
        print(f"{self.color} cat is {self.action}")

    def changeColor(self, color):
        self.color = color

    @classmethod
    def no_parameter(cls):
        return cls("White", "sitting")

    @classmethod
    def first_parameter(cls, color):
        return cls(color, "sitting")

    @classmethod
    def second_parameter(cls, action):
        return cls("Grey", action)

print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
# Task 8
import math

class Cylinder:
    radius = 5
    height = 18
    def __init__(self, radius, height):
        print(f"Default radius={Cylinder.radius} and height={Cylinder.height}.")
        self.radius = radius
        self.height = height
        Cylinder.radius = self.radius
        Cylinder.height = self.height
        print(f"Updated: radius={Cylinder.radius} and height={Cylinder.height}.")

    @staticmethod
    def area(radius, height):
        print("Area:", 2*math.pi*radius**2 + 2*math.pi*radius*height)
    
    @staticmethod
    def volume(radius, height):
        print("Volume:", height*math.pi*radius**2)

    @classmethod
    def swap(cls, radius, height):
        return cls(height, radius)

    @classmethod
    def changeFormat(cls, str1):
        return cls(float(str1.split("-")[0]), float(str1.split("-")[1]))

c1 = Cylinder(0,0) 
Cylinder.area(c1.radius,c1.height) 
Cylinder.volume(c1.radius,c1.height) 
print("===============================") 
c2 = Cylinder.swap(8,3) 
c2.area(c2.radius,c2.height) 
c2.volume(c2.radius,c2.height) 
print("===============================") 
c3 = Cylinder.changeFormat("7-13") 
c3.area(c3.radius,c3.height) 
c3.volume(c3.radius,c3.height) 
print("===============================") 
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height) 
print("===============================") 
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
# Task 10
class SultansDine:
    branch = 0
    sale = 0
    list1 = []

    def __init__(self, loc):
        self.loc = loc
        SultansDine.list1.append(self.loc)
        SultansDine.branch += 1
        

    def sellQuantity(self, sale):
        if sale < 10:
            self.sale = sale * 300
        elif sale < 20:
            self.sale = sale * 350
        else:
            self.sale = sale * 400
        SultansDine.list1.append(self.sale)
        SultansDine.sale += self.sale

    def branchInformation(self):
        print("Branch Name:", self.loc)
        print("Branch Sell:", self.sale, "Taka")

    @classmethod
    def details(cls):
        print("Total Number of branch(s):", SultansDine.branch)
        print("Total Sell:", SultansDine.sale, "Taka")
        for i in range(0, len(SultansDine.list1)-1, 2):
            sale_percent = (SultansDine.list1[i+1]*100/SultansDine.sale)
            rounded_sale = "{:.2f}".format(sale_percent)
            print(f"Branch Name: {SultansDine.list1[i]}, Branch Sell: {SultansDine.list1[i+1]} Taka\nBranch consists of total sell's: {rounded_sale}%")


SultansDine.details() 
print('########################') 
dhanmodi = SultansDine('Dhanmondi') 
dhanmodi.sellQuantity(25) 
dhanmodi.branchInformation() 
print('-----------------------------------------') 
SultansDine.details() 
 
print('========================') 
 
baily_road = SultansDine('Baily Road') 
baily_road.sellQuantity(15) 
baily_road.branchInformation() 
print('-----------------------------------------') 
SultansDine.details() 
 
print('========================') 
 
gulshan = SultansDine('Gulshan') 
gulshan.sellQuantity(9) 
gulshan.branchInformation() 
print('-----------------------------------------') 
SultansDine.details()
```
